```python
# Indexing & Selection
import pandas as pd

data = {
    "Name": ["Laksh", "Aryan", "Parth", "Riya"],
    "Age": [20, 21, 22, 19],
    "Marks": [90, 85, 95, 88],
    "City": ["Mumbai", "Pune", "Delhi", "Mumbai"]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Selection using column names
print(df[["Name"]])
```

```python
# Selection using multiple column names
print(df[["Name", "Marks"]])
```

```python
# loc[] → Label Based Selection
print(df.loc[[0]])  # Selects the first row (index label 0)
print()
print(df.loc[0:2])  # Selects rows from index label 0 to 2 (inclusive)
```

```python
# Select Row + Column
print(df.loc[0, "Name"])
```

```python
# Select Multiple Rows + Columns
print(df.loc[[0,2], ["Name", "Marks"]])
print()
print(df.loc[1:3, ["Name", "Marks"]])
```

```python
# iloc[] → Position Based Selection
print(df.iloc[[0]])  # Selects the first row (position 0)
print() 
print(df.iloc[0:2])  # Selects rows from position 0 to 1 (exclusive of 2)
```

```python
# Select Row and Column using iloc
print(df.iloc[0, 1])  # Selects the value at the first row and first column (position 0,0)

# Explanation:

# Row position = 0
# Column position = 1 (Age)
```

```python
# Select Multiple Rows + Columns using iloc
print(df.iloc[[0,2], [0,2]])
```

Difference Between loc[] and iloc[]
loc[]             |	  iloc[]
                  |
Uses labels	      |   Uses positions
Uses column names |   Uses column numbers

```python
# Examples of loc[] and iloc[] in Pandas:

# loc[] → Label Based Selection
print(df.loc[0, "Name"])
# Uses:
# Row label
# Column name

print()

# iloc[] → Position Based Selection
print(df.iloc[0, 0])
# Uses:
# Row position  
# Column position
```

```python
# Set Index

df.set_index("Name", inplace=True)
print(df)
```

```python
# Acessing Data using Index
print(df.loc[["Laksh"]])
```

```python
# Reset Index
df = df.reset_index()
print(df)
```
