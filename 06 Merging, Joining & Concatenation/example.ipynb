```python
import pandas as pd

students ={
    "ID": [1, 2, 3, 4],
    "Name": ["Laksh", "Aryan", "Parth", "Jyoti"]
}
students  = pd.DataFrame(students)

print(students)
```

```python
marks = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Marks": [90, 85, 95, 98]
})

print(marks)
```

```python
# Merge DataFrames
result = pd.merge(students, marks, on="ID")

print(result)
```

```python
# INNER JOIN (Default)
students = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Name": ["Laksh", "Aryan", "Parth", "Jyoti"]
})

marks = pd.DataFrame({
    "ID": [1, 2],
    "Marks": [90, 85]
})

print(pd.merge(students, marks, on="ID"))
```

```python
# LEFT JOIN
print(
    pd.merge(
        students,
        marks,
        on="ID",
        how="left"
    )
)
```

```python
# RIGHT JOIN
print(
    pd.merge(
        students,
        marks,
        on="ID",
        how="right"
    )
)
```

```python
# OUTER JOIN
print(
    pd.merge(
        students,
        marks,
        on="ID",
        how="outer"
    )
)
```

```python
# Merge on Different Column Names
students = pd.DataFrame({
    "Student_ID": [1, 2],
    "Name": ["Parth", "Jyoti"]
})

marks = pd.DataFrame({
    "ID": [1, 2],
    "Marks": [85, 98]
})

print(
    pd.merge(
        students,
        marks,
        left_on="Student_ID",
        right_on="ID"
    )
)
```

```python
# Multiple Key Merge
df1 = pd.DataFrame({
    "ID": [1, 1],
    "Subject": ["Math", "Science"]
})

df2 = pd.DataFrame({
    "ID": [1, 1],
    "Subject": ["Math", "Science"],
    "Marks": [85, 98]
})

print(
    pd.merge(
        df1,
        df2,
        on=["ID", "Subject"]
    )
)
```

```python
# JOIN()
df1 = pd.DataFrame({
    "Name": ["Laksh", "Aryan", "Parth", "Jyoti"]
}, index=[1, 2, 3, 4])

df2 = pd.DataFrame({
    "Marks": [90, 85, 95, 98]
}, index=[1, 2, 3, 4])

print(df1.join(df2))
```

### Difference Between merge() and join()

| merge() | join() |
|----------|----------|
| Uses columns | Uses indexes |
| More flexible | Simpler |
| SQL-style joins | Index-based joins |

```python
# CONCAT()
df1 = pd.DataFrame({
    "Name": ["Laksh", "Aryan"]
})

df2 = pd.DataFrame({
    "Name": ["Parth", "Jyoti"]
})

print(pd.concat([df1, df2]))
```

```python
# Reset Index
print(
    pd.concat(
        [df1, df2],
        ignore_index=True
    )
)
```

```python
# Horizontal Concatenation
df1 = pd.DataFrame({
    "Name": ["Laksh", "Aryan"]
})

df2 = pd.DataFrame({
    "Marks": [90, 85]
})

print(
    pd.concat(
        [df1, df2],
        axis=1
    )
)
```

Understanding axis
- axis=0: Concatenate along rows (default)
- axis=1: Concatenate along columns
