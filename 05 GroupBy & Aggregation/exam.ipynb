```python
import pandas as pd

data = {
    "Name": ["Laksh", "Aryan", "Parth", "Riya", "Karan"],
    "City": ["Mumbai", "Pune", "Mumbai", "Delhi", "Pune"],
    "Marks": [90, 85, 95, 88, 92],
    "Age": [20, 21, 22, 19, 23]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Grouping by the "City" column
grouped = df.groupby("City")

print(grouped)

print("""
Explanation
- Data is grouped internally.
- No calculation has been performed yet.
""")
```

```python
# Displaying the groups
print(grouped.groups)

print("""
Explanation
- Mumbai group contains rows 0 and 2
- Pune group contains rows 1 and 4
- Delhi group contains row 3
""")
```

```python
# Aggregation Functions

print("1. mean() → Average")
print(df.groupby("City")["Marks"].mean())

print("\n2. sum() → Total")
print(df.groupby("City")["Marks"].sum())

print("\n3. count() → Count of values")
print(df.groupby("City")["Marks"].count())

print("\n4. min() → Minimum value")
print(df.groupby("City")["Marks"].min())

print("\n5. max() → Maximum value")
print(df.groupby("City")["Marks"].max())

print("\n6. median() → Median value")
print(df.groupby("City")["Marks"].median())
```

```python
# Multiple Aggregations
print(
    df.groupby("City")["Marks"]
      .agg(["count","mean", "sum", "max", "min","median"])
)
```

```python
# Group By Multiple Columns

print(df.groupby(["City", "Age"])["Marks"].mean())
print("""
Explanation:
Groups data using both:
- City
- Age
""")
```

```python
# Multiple Columns Aggregation
print(df.groupby("City")[["Marks", "Age"]].mean())
```
