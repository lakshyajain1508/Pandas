```python

# A Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). It is similar to a column in a spreadsheet or a SQL table.

# Creating a Series

# The downward code creates a Series with the values 90, 85, and 95. By default, the index of the Series will be 0, 1, and 2.

import pandas as pd

marks = pd.Series([90, 85, 95])

print(marks)
```

```python
# Creating a Series with custom index
# The downward code creates a Series with the values 90, 85, and 95, and assigns custom index labels 'Math', 'Science', and 'English' to the corresponding values.

import pandas as pd

marks = pd.Series([90, 85, 95], index=['Math', 'Science', 'English']) 

print(marks)
```

```python
# Accessing elements in a Series

# You can access elements in a Series using the index labels or the integer position. The downward code demonstrates both methods of accessing the first element of the Series.
print("Not Define Indexes")
marks = pd.Series([90, 85, 95])
print("Index 0 - " , marks[0])  # Output: 90

print()

# The downward code demonstrates how to access the first element of the Series using .iloc.
print("Define Indexes")
marks = pd.Series([95, 85, 95], index=['Math', 'Science', 'English']) 
print("Maths - " ,marks.iloc[0])  # Output: 95
```

```python
# Accessing Multiple values in a Series
# You can also access multiple values in a Series using a list of index labels or integer positions. The downward code demonstrates both methods of accessing multiple elements from the Series.

print("Not Define Indexes")
marks = pd.Series([90, 85, 95])
print(marks[[0, 2]])
print()
print("Define Indexes")
marks = pd.Series([95, 85, 80], index=['Math', 'Science', 'English'])
print(marks[['Math', 'English']])
```

```python
# Creating a Series from a dictionary
# You can create a Series from a dictionary where the keys become the index labels and the values become the corresponding values in the Series. The downward code demonstrates how to create a Series from a dictionary containing student names as keys and their salary as values.

data = {
    "Laksh": 90000,
    "Aryan": 85000,
    "Parth": 95000
}

salary = pd.Series(data)

print(salary)
```

```python
# You can also check the data type of the Series using the type() function. The downward code demonstrates how to check the data type of the 'salary' Series.
print(type(salary))
```

```python
# Series Attributes

# Index: The index attribute of a Series returns the index labels of the Series. The downward code demonstrates how to access the index labels of the 'salary' Series.
print ("Index of the Series - " , salary.index)

# Values: The values attribute of a Series returns the actual values in the Series. The downward code demonstrates how to access the values of the 'salary' Series.
print ("Values of the Series - " , salary.values)

# Data Type: The dtype attribute of a Series returns the data type of the values in the Series. The downward code demonstrates how to check the data type of the 'salary' Series.
print ("Data type of the Series - " , salary.dtype)
```

```python
# Series Operations: You can perform various operations on Series, such as arithmetic operations, aggregation functions, and more.

numbers = pd.Series([1, 2, 3])

# Addition
print("Addition")
print(numbers + 10)

print()

# Subtraction
print("Subtraction")
print(numbers - 1)

print()
# Multiplication
print("Multiplication")
print(numbers * 2)

print()
# Division
print("Division")
print(numbers / 2)

print()
# Aggregation Functions
print("Aggregation Functions")
print("Sum - " , numbers.sum())
print("Mean - " , numbers.mean())
print("Max - " , numbers.max())
print("Min - " , numbers.min())
print("Standard Deviation - " , numbers.std())
print("Variance - " , numbers.var())
print("Count - " , numbers.count())
print("Median - " , numbers.median())
print("Mode - " , numbers.mode())
print("Unique Values - " , numbers.unique())
print("Value Counts - " , numbers.value_counts())
```

```python
# Filtering: You can filter a Series based on certain conditions. The downward code demonstrates how to filter the 'numbers' Series to get only the values greater than 1.

print(numbers[numbers > 1])
```

```python
# Checking for null values: You can check for null values in a Series using the isnull() method.

data = pd.Series([10, 20, None, 40])
print(data)

print()

# Check for null values
print(data.isnull())
```

```python
# Sorting by Index

data1 = pd.Series([30, 10, 20], index=['c', 'a', 'b'])
print("Original Series")
print(data1)

print()

# Sort by index
data2 = data1.sort_index()
print("Series sorted by index")
print(data2)
```

```python
# Checking Conditions

print(data2)
print(data2 > 15)
```
