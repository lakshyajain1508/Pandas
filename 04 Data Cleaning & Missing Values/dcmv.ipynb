```python
import pandas as pd

data = {
    "Name": ["Laksh", "Aryan", "Laksh", "Parth"],
    "Age": ['20', None, '20', '20'],
    "Marks": [90, 85, 90, None],
    "City": ["Mumbai", "Pune", "Mumbai", None]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Detect Missing Values
print(df.isnull())

print("\nTrue means missing value exists.")
```

```python
# Counting Missing Values
print(df.isnull().sum())
```

```python
# Remove Missing Values
print(df.dropna())
```

```python
# Drop Columns with Missing Values
print(df.dropna(axis=1))
print()
print(df.dropna(axis=1, how='all'))
```

```python
# Fill missing values wirh zero
print(df.fillna(0))
```

```python
# Fill missing values with a specific column
df["Marks"] = df["Marks"].fillna(50)
df["City"] = df["City"].fillna("Pune")
df["Age"] = df["Age"].fillna("20")
print(df)
```

```python
# Fill with Median : Median is the middle value of a sorted list. It is less affected by outliers and skewed data than the mean.

import pandas as pd

data1 = {
    "Name": ["Laksh", "Aryan", "Laksh", "Parth"],
    "Age": ['20', None, '20', '20'],
    "Marks": [90, 85, 90, None],
    "City": ["Mumbai", "Pune", "Mumbai", None]
}

df1 = pd.DataFrame(data1)

df1["Marks"] = df1["Marks"].fillna(df1["Marks"].median())
print(df1)
```

```python
# Fill with Mode : Mode is the most frequently occurring value in a column. It is used to fill missing values in categorical columns.


df1["City"] = df1["City"].fillna(df1["City"].mode()[0])
print(df1)
```

```python
# Forward Fill : This method fills the missing values with the last valid observation. It is useful when the data is ordered and we want to propagate the last known value forward.
print(df1.ffill())
```

```python
# Backward Fill : This method fills the missing values with the next valid observation. It is useful when the data is ordered and we want to propagate the next known value backward. 
print(df1.bfill())
```

```python
# Duplicate Data : Duplicate data can lead to inaccurate analysis and insights. It is important to identify and remove duplicate rows from the dataset.

print(df1.duplicated())
```

```python
# Drop Duplicates : The drop_duplicates() method is used to remove duplicate rows from the DataFrame. By default, it keeps the first occurrence of the duplicate and drops the rest.

print(df.drop_duplicates())
```

```python
# drop Duplicates pemanently
df1.drop_duplicates(inplace=True)
print(df1)
```

```python
# Check Data Types
print(df1.dtypes)
```

```python
# Convert to Integer
df1["Marks"] = df1["Marks"].astype(int)
print(df1.dtypes)
```

```python
# Convert to Float
df1["Marks"] = df1["Marks"].astype(float)
print(df1.dtypes)
```

```python
# Convert to String
df1["Marks"] = df1["Marks"].astype(str)
print(df1.dtypes)
```
