```python
import pandas as pd

data = {
    "Product": ["Laptop", "Mobile", "Laptop", "Tablet", "Mobile"],
    "City": ["Mumbai", "Pune", "Pune", "Mumbai", "Mumbai"],
    "Sales": [50000, 30000, 45000, 20000, 35000],
    "Quantity": [5, 10, 4, 3, 8]
}

df = pd.DataFrame(data)

print(df)
```

```python
pivot = pd.pivot_table(
    df,
    values="Sales",
    index="Product"
)

print(pivot)

print()
```

```python
print(pd.pivot_table
    (
    df,
    values="Quantity",
    index="Product",
    aggfunc="sum"
    )
)
```

```python
pivot = pd.pivot_table(
    df,
    values="Sales",
    index="Product",
    aggfunc=["sum", "mean", "max"]
)

print(pivot)
```

```python
# pivot table with multiple columns
pivot = pd.pivot_table(
    df,
    values="Quantity",
    index="Product",
    aggfunc=["sum", "mean", "max"]
)
print(pivot)
```

```python
# Pivot table with multiple columns
pivot = pd.pivot_table(
    df,
    values="Sales",
    index="Product",
    columns="City",
    aggfunc="sum"
)

print(pivot)
```

```python
# Fill missing values with 0
pivot = pd.pivot_table(
    df,
    values="Sales",
    index="Product",
    columns="City",
    aggfunc="sum",
    fill_value=0
)

print(pivot)
```

```python
# Multiple Value Columns
pivot = pd.pivot_table(
    df,
    values=["Sales", "Quantity"],
    index="Product",
    aggfunc="sum"
)

print(pivot)
```

```python
# Multiple Indexes
pivot = pd.pivot_table(
    df,
    values="Sales",
    index=["City", "Product"],
    aggfunc="sum"
)

print(pivot)
```

```python
# Margins (Grand Total)
pivot = pd.pivot_table(
    df,
    values="Sales",
    index="Product",
    aggfunc="sum",
    margins=True
)

print(pivot)
```
