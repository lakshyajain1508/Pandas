```python
import pandas as pd
data = {
    "Gender": ["Male", "Female", "Male", "Female", "Male"],
    "Result": ["Pass", "Pass", "Fail", "Pass", "Fail"]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Basic Crosstab

print(
    pd.crosstab(
        df["Gender"],
        df["Result"]
    )
)
```

```python
# Add Totals
print(
    pd.crosstab(
        df["Gender"],
        df["Result"],
        margins=True
    )
)
```

```python
# Normalize Crosstab

print(
    pd.crosstab(
        df["Gender"],
        df["Result"],
        normalize=True
    )
)
```

# normalize Options

| Option      | Meaning            |
| ----------- | ------------------ |
| `True`      | Overall percentage |
| `"index"`   | Row percentage     |
| `"columns"` | Column percentage  |

```python
# Row Percentage
print(
    pd.crosstab(
        df["Gender"],
        df["Result"],
        normalize="index"
    )
)
```
